7.Describe attributes of data, checking data types of each column,

```python
import pandas as pd

data = {
    'student_id': [101, 102, 103, 104, 105],
    'name': ['Rushi', 'Ravi', 'Sita', 'Neha', 'Rahul'],
    'age': [20, 21, 22, 20, 23],
    'gender': ['Male', 'Male', 'Female', 'Female', 'Male'],
    'department': ['CS', 'IT', 'CS', 'ENTC', 'MECH'],
    'marks': [78, 85, 0, 92, 88],
    'attendance': [90, 75, 80, 95, 0],
    'internship': ['Yes', 'No', 'Yes', 'Yes', 'No']
}

df = pd.DataFrame(data)

def gap():
  print("\n" + "-"*50 + "\n")

print("1) Shape of Data (Rows, Columns)")
print(df.shape)
gap()


print("2) Missing Values in each column")
print(df.isnull().sum())
gap()


print("3) Data Types of each column")
print(df.dtypes)
gap()


print("4) Zero values in each column")
print((df == 0).sum())
gap()


print("5) Indexing & Selecting Data")

print("\n-- Name Column --")
print(df['name'])

print("\n-- First 3 Rows --")
print(df.iloc[0:3])

print("\n-- Students with Marks > 80 --")
print(df[df['marks'] > 80])
gap()


print("6) Sorting Data by Marks (Descending)")
print(df.sort_values(by='marks', ascending=False))
gap()

print("6) Sorting Data by Marks (Aescending)")
print(df.sort_values(by='marks', ascending=True))
gap()

print("7) Describe Data")
print(df.describe())
gap()


print("8) DataFrame Info")
df.info()
gap()


print("9) Unique values count in each column")
print(df.nunique())
gap()


print("10) Average Marks Department wise")
print(df.groupby('department')['marks'].mean())
gap()


print("11) Data Type Conversion")

df['student_id'] = df['student_id'].astype(str)
df['marks'] = df['marks'].astype(float)
df['internship'] = df['internship'].map({'Yes': True, 'No': False})

print(df.dtypes)
gap()


print("12) Convert String Columns to Lowercase")
df = df.apply(lambda col: col.str.lower() if col.dtype == 'object' else col)
print(df)
gap()

print(" ALL OPERATIONS COMPLETED SUCCESSFULLY")
```

# Merge:
pandas provides various facilities for easily combining together Series and DataFrame objects with various kinds of set logic for the indexes and relational algebra functionality in the case of join / merge-type operations.

See the Merging section.

Concatenating pandas objects together row-wise with concat():

```python
import pandas as pd

df1 = df.iloc[0:3]   # first 3 rows
df2 = df.iloc[3:5]   # last 2 rows

result = pd.concat([df1, df2])
print(result)
```
