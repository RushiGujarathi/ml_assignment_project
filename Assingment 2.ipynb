user-defined function for varience

```python
def user_variance(dat):
    n = len(dat)
    mean = sum(dat) / n
    var = sum((x - mean) ** 2 for x in dat) / (n - 1)
    return var

dat=[10,20,30,40,50]
print(user_variance(dat))
```

 user-defined function for varience for mean ,median ,mode.

```python

def user_mean(dat):
    return sum(dat) / len(dat)



def user_median(dat):
    dat = sorted(dat)
    n = len(dat)
    mid = n // 2

    if n % 2 == 0:
        return (dat[mid - 1] + dat[mid]) / 2
    else:
        return dat[mid]



def user_mode(dat):
    freq = {}
    for x in dat:
        freq[x] = freq.get(x, 0) + 1

    max_freq = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_freq]

    return modes if len(modes) > 1 else modes[0]



dat = [10, 20, 30, 40, 50]

print("Mean:", user_mean(dat))
print("Median:", user_median(dat))
print("Mode:", user_mode(dat))
```

# **Data Integration**

Vertical Data Integration (Row-wise) Combines datasets with same structure by stacking rows.

```python
import pandas as pd

jan = pd.DataFrame({
    'EmpID': [1, 2],
    'Sales': [20000, 25000]
})

feb = pd.DataFrame({
    'EmpID': [3, 4],
    'Sales': [22000, 27000]
})

Mar=pd.DataFrame({
    'EmpID' :[5,6],
    'Sales' :[45000,3700]

})
# Integrate data row-wise
sales_data = pd.concat([jan, feb,Mar], ignore_index=True)

print(sales_data)
```

Horizontal Data Integration (Column-wise) Combines different attributes for the same entities.
Student Academic & Attendance Data Using concat(axis=1)

```python
marks = pd.DataFrame({
    'RollNo': [101, 102, 103],
    'Marks': [85, 90, 88]
})

attendance = pd.DataFrame({
    'RollNo': [101, 102, 103],
    'Attendance': [92, 95, 90]
})


# Integrate data column-wise
student_data = pd.concat([marks, attendance['Attendance']], axis=1)

print(student_data)
```

Key-based Data Integration (Joins)  Using merge()                      Integrates data based on a common key.


```python
students = pd.DataFrame({
    'StudentID': [1, 2, 3],
    'Name': ['Amit', 'Neha', 'Rahul']
})

results = pd.DataFrame({
    'StudentID': [1, 2],
    'CGPA': [8.5, 9.1]
})

# Inner Join
data_inner = pd.merge(students, results, on='StudentID', how='inner')

print(data_inner)
```

Types of Joins in Data Integration

```python
print(pd.merge(students, results, on='StudentID', how='inner') )  # Common records
print(pd.merge(students, results, on='StudentID', how='left'))    # All left + matches
print(pd.merge(students, results, on='StudentID', how='right') )  # All right + matches
print(pd.merge(students, results, on='StudentID', how='outer')  ) # All records
```

Integration Using Index Using join() Joins datasets using index instead of columns.

```python
df1 = students.set_index('StudentID')
df2 = results.set_index('StudentID')

integrated_data = df1.join(df2)

print(integrated_data)
```

# Data transformation

1.MinMaxScaler vs StandardScaler
Both are feature scaling techniques, but they work differently.

A. MinMaxScaler

Purpose: Scales all numeric values to a fixed range, usually 0 to 1 (can be changed). Formula:

𝑋 scaled=𝑋−𝑋min/𝑋max −𝑋 min

⁡ X_min = minimum value of feature

X_max = maximum value of feature

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

df = pd.DataFrame({'Age':[18,22,25,30,35]})
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
print(df_scaled)
```

When to use:

Data is not normally distributed

You need fixed range for all features

Algorithms like Neural Networks, KNN, SVM are sensitive to scale

A. Scaling / Normalization / Standardization

Purpose: Make numerical features comparable.

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd

df = pd.DataFrame({'Age':[18,22,25,30], 'Salary':[25000,50000,60000,80000]})


scaler = MinMaxScaler()
df[['Age','Salary']] = scaler.fit_transform(df[['Age','Salary']])
print("Min-Max Scaled:\n", df)


scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['Age','Salary']])
print("Standardized:\n", df_scaled)
```
