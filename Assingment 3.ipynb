Visualize the data using Python by plotting the graphs for assignment no. 1 and 2. Consider a suitable data set. Use Scatter plot, Bar plot, Box plot, Pie chart, Line Chart.

```python
import pandas as pd
data={
    'student':['Rushi','Aayush','Rahul','omkar','om'],
    'Maths':[90,95,99,97,89],
    'Science':[88,99,77,88,99],
    'English':[92,88,90,98,99]
}

df=pd.DataFrame(data);
print(df)
```

Scatter Plot

Purpose: Relationship between two numerical variables (Maths vs Science)

```python
import matplotlib.pyplot as plt

plt.scatter(df['Maths'], df['Science'])
plt.xlabel('Maths Marks')
plt.ylabel('Science Marks')
plt.title('Scatter Plot: Maths vs Science')
plt.show()
```

Bar Plot

Purpose: Compare marks of students in ENGLISH subject

```python
plt.bar(df['student'], df['English'])
plt.xlabel("Student name")
plt.ylabel("English marks")
plt.title("student marks")
plt.show()
```

Box Plot

Purpose: Show distribution and outliers of marks

```python
plt.boxplot([df['Maths'], df['Science'], df['English']],
            tick_labels=['Maths', 'Science', 'English'])
plt.title('Box plot diagram')
plt.ylabel("Marks of the student")


plt.show()
```

Pie Chart

Purpose: Percentage contribution of total marks by subject (average

```python
average_marks=df[['Maths','Science','English']].mean()

plt.pie(average_marks,labels=average_marks.index,autopct='%1.1f%%')
plt.title("Pie Chart: Average Marks Distribution")
plt.show()
```

Line Chart

Purpose: Trend of marks across students

```python
import matplotlib.pyplot as plt
plt.plot(df['student'], df['Maths'], label='Maths')
plt.plot(df['student'], df['Science'], label='Science')
plt.plot(df['student'], df['English'], label='English')

plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Line Chart: Marks Trend')
plt.legend()
plt.show()
```

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

df['hour'] = df['pickup_datetime'].dt.hour
df['day'] = df['pickup_datetime'].dt.day
df['month'] = df['pickup_datetime'].dt.month

df.drop('pickup_datetime', axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['fare_amount'])
plt.ylabel('Fare Amount')
plt.title('Box Plot: Fare Amount (Outlier Detection)')
plt.show()
```

Fare Distribution – Histogram

```python
plt.hist(df['fare_amount'], bins=50)
plt.xlabel('Fare Amount')
plt.ylabel('Frequency')
plt.title('Histogram: Fare Amount Distribution')
plt.show()
```

Correlation Matrix – Heatmap (Matplotlib)

```python
corr = df.corr()

plt.matshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('Correlation Matrix')
plt.show()
```
